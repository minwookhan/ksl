```python
import csv
with open('gloss_for_keyframe_20230808r3.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    gloss_list = []
    for row in csvreader:
        if 'o' in row[5]:
            gloss_list.append(int(row[1]))
            # print(row)
    print(gloss_list)
```

```python
import os
gloss_folder = 'C:/Work/sign_gloss'

gfile_list = []
file_list = os.listdir(gloss_folder)
for gloss in gloss_list:
    cand_gfile_name1 = f'{gloss}.001.C.json'
    cand_gfile_name2 = f'{gloss}.002.C.json'
    if cand_gfile_name1 in file_list:
        gfile_list.append(cand_gfile_name1)
    if cand_gfile_name2 in file_list:
        gfile_list.append(cand_gfile_name2)
print(len(gfile_list), gfile_list)
```

```python
import changepointDetect
import scipy.ndimage
import numpy as np
from thresholds import *
from matplotlib import pyplot as plt

i = 0
for gfilename in gfile_list:
    gfile_path = os.path.join(gloss_folder, gfilename)
    # print(gfile_path)
    rHandtip, lHandtip = changepointDetect.getHandTipJointData(gfile_path)

    strong_cp, weak_cp = changepointDetect.changepoint_detector(rHandtip, min_threshold=MIN_THRESHOLD, minor_threshold=MINOR_THRESHOLD, 
        adj_local_threshold=ADJ_LOCAL_THRESHOLD, sigma=SIGMA, minor_max_threshold=MINOR_MAX_THRESHOLD, angular_acc_threshold=ANGULAR_ACC_THRESHOLD)

    print(gfilename)
    print(f'New Strong Changepoints: {strong_cp}, New Weak Changepoints: {weak_cp}')

    cp_array = changepointDetect.changepoint_detector_1array_confi(rHandtip, min_threshold=MIN_THRESHOLD, minor_threshold=MINOR_THRESHOLD, 
        adj_local_threshold=ADJ_LOCAL_THRESHOLD, sigma=SIGMA, minor_max_threshold=MINOR_MAX_THRESHOLD, angular_acc_threshold=ANGULAR_ACC_THRESHOLD)
    print(f'Array: {cp_array}')

    handtip_gfiltered = scipy.ndimage.gaussian_filter1d(rHandtip, SIGMA, axis=0)
    handtipDisp = np.array(changepointDetect.calculateDisplacement(handtip_gfiltered))
    handtipDispGaussFiltered = scipy.ndimage.gaussian_filter1d(handtipDisp, SIGMA)
    handtipDispVector = changepointDetect.calculateDisplacementVector(handtip_gfiltered)
    angAcc_org = changepointDetect.calculateAngularAcc(handtipDispVector)
    angAcc_gfiltered = scipy.ndimage.gaussian_filter1d(angAcc_org, SIGMA)


    plt.rcParams["figure.figsize"] = (15,5)
    plt.subplot(2,2,1)
    plt.plot(handtipDispGaussFiltered, label="Right Handtip Displacement")
    plt.plot(strong_cp, handtipDispGaussFiltered[strong_cp], 'bx')
    plt.plot(weak_cp, handtipDispGaussFiltered[weak_cp], 'gx')
    plt.grid(True)
    # plt.legend()
    plt.subplot(2,2,2)
    plt.plot(angAcc_gfiltered, label="Angular Acceleration")
    plt.plot(strong_cp, angAcc_gfiltered[strong_cp], 'bx')
    plt.plot(weak_cp, angAcc_gfiltered[weak_cp], 'gx')
    plt.grid(True)
    # plt.legend()
    plt.show()

    i += 1
    # if i > 3:
    #     break
```
